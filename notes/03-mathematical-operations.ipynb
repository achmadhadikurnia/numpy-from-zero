# 3. Mathematical Operations

NumPy makes mathematical operations over large datasets incredibly fast and concise compared to standard Python loops.

We will cover:
1. **Element-wise Operations**: Basic arithmetic (+, -, *, /) applied to every element.
2. **Aggregation Functions**: Summarizing data using `sum`, `mean`, `max`, `min`, and `std`.
3. **Matrix Operations & Linear Algebra**: Mathematical operations commonly used in linear algebra, such as dot products, inverse, and determinants.
4. **Universal Math Functions (ufuncs)**: Built-in fast mathematical functions.
5. **Rounding**: Techniques for rounding arrays.
6. **Argmax and Argmin**: Finding the position (index) of maximum and minimum values.
7. **Advanced Statistics**: Calculating medians and percentiles.
8. **Absolute Values**: Using `np.abs`.
9. **Value Clipping**: Restricting values to a range using `np.clip`.
10. **Cumulative Computations**: Calculating cumulative sum with `np.cumsum`.
11. **Array Comparison**: Comparing arrays with `np.allclose`, `np.array_equal`, and `np.isclose`.

```python
import numpy as np
```

## 3.1 Element-wise Operations (Arithmetic)
Arithmetic operations on arrays apply to each element individually. This is highly useful for tasks like **feature scaling** or applying mathematical transformations to an entire dataset at once (e.g., converting temperatures from Celsius to Fahrenheit without a `for loop`).

```python
# 1. Basic Arithmetic
arr1 = np.array([10, 20, 30])
arr2 = np.array([1, 2, 3])

print("Addition (+):", arr1 + arr2)
print("Subtraction (-):", arr1 - arr2)
print("Multiplication (*):", arr1 * arr2)
print("Division (/):", arr1 / arr2)

# 2. Operations with Scalars (Single numbers)
# Useful for applying a constant transformation, like adding a bias.
print("\nAdd 5 to everything:", arr1 + 5)
print("Square everything (**2):", arr1 ** 2)
```

## 3.2 Aggregation Functions
Aggregation functions summarize data into a single value. These are essential for **statistical analysis**, finding the average of a metric, finding the maximum/minimum bounds, or calculating the standard deviation (spread of data).

```python
data = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print("Data:\n", data)
print("\nSum of all elements:", np.sum(data))
print("Mean (Average):", np.mean(data))
print("Maximum value:", np.max(data))
print("Minimum value:", np.min(data))
print("Standard Deviation:", np.std(data))

# VERY IMPORTANT: Aggregating along an axis
# axis=0 aggregates column by column (e.g., finding the average of each feature/column)
# axis=1 aggregates row by row (e.g., finding the total score of each student/row)
print("\nSum of each column (axis=0):", np.sum(data, axis=0))
print("Sum of each row (axis=1):", np.sum(data, axis=1))
```

## 3.3 Matrix Operations & Linear Algebra
These operations follow the rules of linear algebra. The **Dot Product** is particularly fundamental in **Machine Learning**, especially in Neural Networks for calculating weights and inputs. We also introduce basic linear algebra like finding the inverse and determinant.

```python
matrix_A = np.array([[1, 2],
                     [3, 4]])

matrix_B = np.array([[5, 6],
                     [7, 8]])

# 1. Element-wise multiplication (NOT a dot product)
print("Element-wise multiplication:\n", matrix_A * matrix_B)

# 2. Dot Product (Matrix Multiplication)
# This is heavily used in Deep Learning layers.
dot_product = np.dot(matrix_A, matrix_B)
# Alternatively: dot_product = matrix_A @ matrix_B
print("\nDot Product:\n", dot_product)

# 3. Transpose
# Swaps rows and columns. Useful when dimensions need to align for a dot product.
print("\nTranspose of A:\n", matrix_A.T)

# 4. Inverse and Determinant (Linear Algebra)
square_matrix = np.array([[1, 2], [3, 4]])
inverse_matrix = np.linalg.inv(square_matrix)
determinant = np.linalg.det(square_matrix)
print("\nInverse of square_matrix:\n", inverse_matrix)
print("\nDeterminant of square_matrix:", determinant)
```

## 3.4 Universal Math Functions (ufuncs)
NumPy provides many universal functions that apply mathematical operations element-by-element. These are widely used for **data transformations**.

```python
arr_math = np.array([1, 4, 9, 16])

print("Original array:", arr_math)
print("Square Root (np.sqrt):", np.sqrt(arr_math))

# Exponential (e^x) and Logarithm
# Highly used in statistical formulas and machine learning activation functions
arr_small = np.array([1, 2, 3])
print("\nExponential (np.exp):", np.exp(arr_small))
print("Natural Log (np.log):", np.log(arr_small))
```

## 3.5 Rounding
When dealing with floating-point numbers, you often need to round them to integers or a specific number of decimal places.

```python
arr_floats = np.array([1.2, 2.5, 3.8, -1.2, -2.8])
print("Original floats:", arr_floats)

print("\nStandard Rounding (np.round):", np.round(arr_floats))
print("Round Down (np.floor):", np.floor(arr_floats))
print("Round Up (np.ceil):", np.ceil(arr_floats))
```

## 3.6 Argmax and Argmin
While `max()` and `min()` tell you *what* the largest/smallest value is, `argmax()` and `argmin()` tell you **where** (at what index) that value is located. This is exceptionally useful in Classification tasks to find the predicted class.

```python
# Imagine these are probabilities output by a Machine Learning model for 3 classes
probabilities = np.array([0.1, 0.7, 0.2])
print("Probabilities:", probabilities)

# --- np.argmax: Which class has the HIGHEST probability? ---
max_val = np.max(probabilities)
predicted_class_index = np.argmax(probabilities)
print(f"\nHighest probability is {max_val}")
print(f"Located at index (Predicted Class): {predicted_class_index}")

# --- np.argmin: Which class has the LOWEST probability? ---
min_val = np.min(probabilities)
least_likely_index = np.argmin(probabilities)
print(f"\nLowest probability is {min_val}")
print(f"Located at index (Least Likely Class): {least_likely_index}")

# --- Using argmin with axis (2D arrays) ---
# This is very useful for finding nearest neighbors (closest distance per row)
distances = np.array([[10, 5, 8],
                      [3, 12, 7],
                      [9, 6, 1]])
print("\nDistance Matrix:\n", distances)
print("Closest neighbor per row (argmin axis=1):", np.argmin(distances, axis=1))
print("Closest neighbor per col (argmin axis=0):", np.argmin(distances, axis=0))
```

## 3.7 Advanced Statistics (Median & Percentiles)
While `mean` is useful, it is highly sensitive to outliers. `median` (the middle value) and `percentile` are robust statistical tools heavily used in Data Analysis.

```python
# Example: Salaries in a company (one CEO earns 1,000,000, others earn normal wages)
salaries = np.array([30000, 35000, 40000, 42000, 1000000])

print("Mean (Average) Salary:", np.mean(salaries)) # Heavily skewed by the CEO!
print("Median (Middle) Salary:", np.median(salaries)) # A much better representation

# Percentiles: 25th, 50th (median), and 75th percentiles
q1, median_val, q3 = np.percentile(salaries, [25, 50, 75])
print(f"\n25th Percentile (Q1): {q1}")
print(f"75th Percentile (Q3): {q3}")
```

## 3.8 Absolute Values (`np.abs`)
Returns the absolute (positive) value of elements. This is fundamentally used when calculating "distance" or "error" between predictions and actual values (e.g., Mean Absolute Error).

```python
actual_prices = np.array([100, 150, 200])
predicted_prices = np.array([110, 140, 250])

# Calculate the raw errors (can be negative)
errors = actual_prices - predicted_prices
print("Raw Errors:", errors)

# Absolute errors (always positive)
abs_errors = np.abs(errors)
print("Absolute Errors:", abs_errors)

print(f"\nMean Absolute Error (MAE): {np.mean(abs_errors)}")
```

## 3.9 Value Clipping (`np.clip`)
`np.clip(array, min_val, max_val)` forces all values in an array to stay within a defined range. Anything below `min_val` becomes `min_val`, and anything above `max_val` becomes `max_val`.

```python
# Imagine pixel values that must remain between 0 and 255
pixels = np.array([-50, 100, 200, 300])
print("Original pixels:", pixels)

# Clip the values to a valid color range [0, 255]
clipped_pixels = np.clip(pixels, 0, 255)
print("Clipped pixels:", clipped_pixels)
```

## 3.10 Cumulative Computations (`np.cumsum`)
`np.cumsum` returns the cumulative sum (running total) of elements along a given axis.

```python
# Daily sales revenue over a 5-day week
daily_sales = np.array([100, 250, 150, 300, 200])
print("Daily Sales:", daily_sales)

# Calculate total revenue at the end of each day
running_total = np.cumsum(daily_sales)
print("Running Total (Cumulative):", running_total)
```

## 3.11 Array Comparison

When working with iterative algorithms (e.g., optimization, clustering), you often need to check if two arrays are **equal** or **close enough** to stop iterating. NumPy provides several functions for this:

```python
# --- np.array_equal ---
# Checks if two arrays have the EXACT same shape and elements.
arr_a = np.array([1, 2, 3])
arr_b = np.array([1, 2, 3])
arr_c = np.array([1, 2, 4])

print("arr_a:", arr_a)
print("arr_b:", arr_b)
print("arr_c:", arr_c)
print("\nnp.array_equal(arr_a, arr_b):", np.array_equal(arr_a, arr_b))  # True
print("np.array_equal(arr_a, arr_c):", np.array_equal(arr_a, arr_c))  # False

# --- np.allclose ---
# Checks if all elements are equal WITHIN A TOLERANCE.
# This is essential when dealing with floating-point numbers,
# where exact equality almost never works due to rounding errors.
old_centroids = np.array([1.0, 2.0, 3.0])
new_centroids = np.array([1.0000001, 2.0000002, 2.9999999])

print("\nold_centroids:", old_centroids)
print("new_centroids:", new_centroids)
print("Exact equal (==):", np.all(old_centroids == new_centroids))         # False!
print("np.allclose()   :", np.allclose(old_centroids, new_centroids))  # True (within tolerance)

# --- np.isclose ---
# Like np.allclose, but returns a boolean array per element instead of a single True/False.
arr_x = np.array([1.0, 2.0, 3.0])
arr_y = np.array([1.0, 2.1, 3.0000001])

print("\nnp.isclose(arr_x, arr_y):", np.isclose(arr_x, arr_y))
# Element 0: True (exact), Element 1: False (too far), Element 2: True (within tolerance)
```
